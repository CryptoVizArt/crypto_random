```python
import pandas as pd
import os

# Set the folder path
folder_path = 'daily_sorted_mcap_125_assets'  # Replace with your actual folder path

# Initialize an empty DataFrame to store the aggregated data
aggregated_data = pd.DataFrame({'rank': range(1, 126)})

# Loop through each CSV file in the folder
for file_name in sorted(os.listdir(folder_path)):
    if file_name.endswith('.csv'):
        # Read the CSV file
        file_path = os.path.join(folder_path, file_name)
        data = pd.read_csv(file_path)
        
        # Extract the 'ID' column
        ids = data['ID'].head(125)
        
        # Pad the column with NaN if it has fewer than 125 entries
        ids = ids.reindex(range(125), fill_value=None)
        
        # Add the extracted 'ID' column to the aggregated DataFrame
        column_name = file_name.split('.')[0]  # Use the file name (without extension) as the column name
        aggregated_data[column_name] = ids.values

# Save the aggregated data to a new CSV file
output_file = 'aggregated_cryptocurrency_ids.csv'
aggregated_data.to_csv(output_file, index=False)

print(f"Aggregated data saved to {output_file}")
```
